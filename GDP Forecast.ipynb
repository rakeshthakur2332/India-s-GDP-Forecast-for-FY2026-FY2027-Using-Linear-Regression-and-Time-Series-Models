```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_excel(r"C:\Users\Rakesh Thakur\Data Science\Forecast.xlsx")
df
```

```python
df["GDP Growth"] = df["GDP Growth"] * 100
df
```

```python
#Linear Regression Model
X = df[["Year"]]
Y = df["GDP Growth"]

model = LinearRegression()
model.fit(X,Y)

df["Predicted"] = model.predict(X)
```

```python
next_year = df["Year"].max() + 1
second_next_year = df["Year"].max() + 2

forecast_next_year = model.predict([[next_year]])[0]
forecast_second_next_year = model.predict([[second_next_year]])[0]

print("Next Year Forecast", next_year, round(forecast_next_year, 2))
print("Second Next Year Forecast", second_next_year, round(forecast_second_next_year, 2))
```

```python
#ARIMA(1,1,0) Explained
# 1 (p=1) → The model looks at one past value to predict the next.
# 1 (d=1) → The series is differenced once (subtracting each value from the previous one) to remove trends.
# 0 (q=0) → No moving average terms are used (no reliance on past errors).
# So ARIMA(1,1,0) is basically:
# Y_t-Y_{t-1}=c+\phi _1(Y_{t-1}-Y_{t-2})+\varepsilon _t
```

```python
from statsmodels.tsa.arima.model import ARIMA

time_series = df["GDP Growth"]

model = ARIMA(time_series, order=(1,1,0))
model_fit = model.fit()

forecast = model_fit.forecast(steps=2)

next_year = df["Year"].max() + 1
second_next_year = df["Year"].max() + 2

forecast_next_year = forecast.iloc[0]
forecast_second_next_year = forecast.iloc[1]

print("Next Year Forecast", next_year, round(forecast_next_year, 2))
print("Second Next Year Forecast", second_next_year, round(forecast_second_next_year, 2))
```
